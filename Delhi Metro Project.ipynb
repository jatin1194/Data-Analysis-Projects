```python
# Creating a Dictionary to map metro line name to color
lines_color ={
    'Red line':'red',
    'Yellow line':'yellow',
    'Blue line':'blue',
    'Blue line branch':'navy',
    'Green line':'green',
    'Green line branch':'forestgreen',
    'Rapid Metro':'silver',
    'Magenta line':'magenta',
    'Voilet line':'purple',
    'Pink line':'pink',
    'Aqua line':'skyblue',
    'Gray line':'grey',
    'Orange line':'orange'
    # can Add more lines and colors as needed
}

# Calculate value counts for each Metro Line
line_counts=df['Metro Line'].value_counts()

# Create a pie chart using custom colors
plt.pie(line_counts,labels=line_counts.index,autopct='%1.1f%%', colors=[lines_color[line] for line in line_counts.index])


# Set aspect ratio to be equal, so the pie is drawn as a circle
plt.axis('equal')

# Add a title
plt.title('Metro Line Distribution')

# Display the pie chart
plt.show()
line_counts
```

```python
pd.crosstab(df['Layout'],df['Metro Line'])
```

```python
# Create a cross-tabulation of Layouts and Metro Lines

layout_by_line=pd.crosstab(df['Layout'],df['Metro Line'])

# Define line colors according to the pattern
line_colors = {
    'Red line': 'red',
    'Yellow line': 'yellow',
    'Blue line': 'blue',
    'Blue line branch': 'blue',
    'Green line branch': 'green',
    'Green line': 'green',
    'Rapid Metro': 'silver',
    'Voilet line': 'purple',  
    'Magenta line': 'magenta',
    'Pink line': 'pink',
    'Aqua line': 'skyblue',
    'Gray line': 'gray',
    'Orange line': 'orange'
}

# Create a list of colors corresponding to each line
colors=[line_colors[line] for line in layout_by_line.columns]


# Plot a stacked bar plot with specified colors
layout_by_line.plot(kind='bar',stacked=True,figsize=(10,6),color=colors)

plt.xlabel('Layout', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.title('Layout Distribution by Metro Line', fontsize=16)
plt.legend(title='Metro Line', loc='upper left',bbox_to_anchor=(1,1))
plt.tight_layout()
plt.show()
```

#  Conclusion

The exploratory analysis of the Delhi Metro network reveals the strategic and phased expansion of one of India's most extensive urban transit systems. Through spatial visualizations, temporal trends, and inter-station distance analysis, we observed how the network has evolved from a compact core to a vast, multi-line system connecting all corners of the National Capital Region.

Key insights include:

    The Blue and Yellow lines form the backbone of the network, supporting major commuter traffic.

    Newer lines like Magenta and Pink focus on enhancing intra-city connectivity and reducing load on older interchanges.

    Interchange stations such as Rajiv Chowk, Kashmere Gate, and Central Secretariat act as vital transit hubs with high degrees of connectivity.

    The distribution of inter-station distances indicates thoughtful planningdenser spacing in urban centers and broader spacing in peripheral areas.

Overall, the Delhi Metro demonstrates a well-structured transit model with potential for smart expansion under Phase IV. This analysis not only provides a deeper understanding of the systems design but also supports future planning in terms of accessibility, efficiency, and urban development.
